## 列名を整形

```python
#カラム抽出
col_data = pd.read_excel('Yachin.xlsx',skiprows=5, usecols="F, H:V", header=None)
col_data = col_data.head(6)
col_data = col_data.drop(col_data.index[[2]])
col_data
```

```python
cols =[]
for i in col_data.columns:
    tg_col = '_'.join(list(col_data[i].dropna()))
    cols.append(tg_col)
cols
```

### データの列名を整形した列名に変更

```python
df.columns = cols
df.head()
```

## データから「総数」の行だけを抽出

```python
cols[1]
```

```python
df2 = df.copy()
df2[cols[1]]
```

```python
df2.query('"住宅の所有の関係" == "00_総数"')
```

```python
df_sosu = df2[df2['住宅の所有の関係'] == '00_総数']
df_sosu
```

```python
df_sosu.columns
```

```python
# データから必要な列を抽出
df_sosu2 = df_sosu[['地域区分－全国・都道府県・市区町村','延べ面積１ｍ2当たり家賃_住宅の家賃の平均_2_家賃０円を含まない_円']]
df_sosu2
```

```python
## エリアコードを分割
df_sosu2['area_code'] = df_sosu2['地域区分－全国・都道府県・市区町村'].str[:5]
```

```python
df_sosu2['地域区分－全国・都道府県・市区町村'].str[:5]
```

```python
## エリアコードを分割
df_sosu2['area_code'] = df_sosu2.loc['地域区分－全国・都道府県・市区町村'].str[:5]
```

```python
df_sosu2['地域区分－全国・都道府県・市区町村'].str.split('_', expand=True)
```

```python
df_sosu2[['area_code','area_name']] = df_sosu2['地域区分－全国・都道府県・市区町村'].copy().str.split('_', expand=True)
```

```python
df_sosu2
```

```python
#Warningが出ないようにデータフレームを作る
df_sosu3 = df_sosu2.copy()
df_sosu3[['area_code','area_name']] = df_sosu2['地域区分－全国・都道府県・市区町村'].str.split('_', expand=True)
df_sosu3
```

# 市区町村の行だけを抽出する

```python
#「地域コードの下二桁00以外のものが市区町村に該当する」とのことなので、、、
df_sosu3['area_flag'] = df_sosu3['area_code'].str[3:5]
```
